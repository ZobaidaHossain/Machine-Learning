# 1.Load heart disease dataset in pandas dataframe

```python
import pandas as pd
df=pd.read_csv("heart.csv")
df.head()
```

# 3.Convert text columns to numbers using label encoding and one hot encoding

```python
subdf=pd.get_dummies(df)
subdf.head()
```

```python
x = subdf.drop("HeartDisease",axis='columns')
y = subdf.HeartDisease

x.head()
```

# 4.Apply scaling

```python
# scaling
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()
X_scaled = scaler.fit_transform(x)
X_scaled
```

# 5.Build a classification model using various methods (SVM, logistic regression, random forest) and check which model gives you the best accuracy

```python
from sklearn.model_selection import train_test_split

x_train, x_test, y_train, y_test = train_test_split(X_scaled, y, test_size=0.2, random_state=30)
```

```python
from sklearn.svm import SVC
model = SVC()
model.fit(x_train,y_train)
model.score(x_test,y_test)
```

```python
from sklearn.ensemble import RandomForestClassifier
model = RandomForestClassifier(n_estimators=20)
model.fit(x_train, y_train)
model.score(x_test, y_test)
```

```python
from sklearn.linear_model import LogisticRegression
model=LogisticRegression()
model.fit(x_train,y_train)
model.score(x_test,y_test)
```

# 6.Now use PCA 

```python
#PCA
from sklearn.decomposition import PCA
pca = PCA(0.95)
x_pca = pca.fit_transform(x)
x_pca
```

```python
x_pca.shape
```

```python
x_train_pca, x_test_pca, y_train, y_test = train_test_split(x_pca, y, test_size=0.2, random_state=30)
```

```python
from sklearn.ensemble import RandomForestClassifier
model = RandomForestClassifier(n_estimators=20)
model.fit(x_train_pca, y_train)
model.score(x_test_pca, y_test)
```
